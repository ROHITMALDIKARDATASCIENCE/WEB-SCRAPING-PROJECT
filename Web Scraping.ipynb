```python
import requests
from bs4 import BeautifulSoup
import pandas as pdbbb
```

```python
page = requests.get("https://zeelretail.com/collections/men")
soup = BeautifulSoup(page.content, 'html.parser')
soup
```

```python
soup.prettify()
```

```python
period = soup.select("product-price__item product-price__from")
# for i in range(0,len(period[1].get_text()
list1=[]
for i in range(0,len(period)):
    list1.append(period[i].get_text())

list1
```

```python
period_tags = soup.select("product-price__item.product-price__from")
period_tags
```

```python
week_period=[]
for i in period_tags:
    week_period.append(i.get_text())

print(week_period)
```
